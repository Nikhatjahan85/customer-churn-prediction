```python
import pandas as pd
import os

# Correct path
file_path = "../data/churn.csv"

# Check file exists
if not os.path.exists(file_path):
    print("❌ File not found:", file_path)
else:
    print("✅ File found!")

    df = pd.read_csv(file_path)

    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())

    if df.shape[0] == 0:
        print("❌ CSV is EMPTY")
    else:
        print(df.head())


import matplotlib.pyplot as plt
import seaborn as sns
import os

# create folder if not exists
os.makedirs("../outputs", exist_ok=True)

sns.set(style="whitegrid")

# ---------------------------
# 1. Churn Distribution
# ---------------------------
plt.figure(figsize=(6,4))
sns.countplot(x=df['Churn'])
plt.title("Churn Distribution")
plt.savefig("../outputs/churn_distribution.png")
plt.show()

# ---------------------------
# 2. Tenure vs Churn
# ---------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Churn'], y=df['tenure'])
plt.title("Tenure vs Churn")
plt.savefig("../outputs/tenure_vs_churn.png")
plt.show()

# ---------------------------
# 3. Monthly Charges vs Churn
# ---------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Churn'], y=df['MonthlyCharges'])
plt.title("Monthly Charges vs Churn")
plt.savefig("../outputs/monthly_vs_churn.png")
plt.show()

# ---------------------------
# 4. Total Charges vs Churn
# ---------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Churn'], y=df['TotalCharges'])
plt.title("Total Charges vs Churn")
plt.savefig("../outputs/total_vs_churn.png")
plt.show()

# ---------------------------
# 5. Correlation Heatmap
# ---------------------------
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.savefig("../outputs/correlation.png")
plt.show()
       
```
